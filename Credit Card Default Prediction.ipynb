### Data Source
This dataset was obtained from Kaggle (UCI Credit Card Default Dataset) and is used for academic and learning purposes only.

```python
import pandas as pd
import numpy as np

data = pd.read_csv("UCI_Credit_Card.csv.zip", header=0)
```

```python
data
```

```python
data.drop("ID", axis=1, inplace=True)
data = data.apply(pd.to_numeric, errors='coerce')
data
```

```python
data.rename(columns={'default.payment.next.month': 'DEFAULT'}, inplace=True)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
X = data.drop('DEFAULT', axis=1)
y = data['DEFAULT']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

categorical_features = ["SEX", "EDUCATION", "MARRIAGE"]
X_encoded = pd.get_dummies(X, columns=categorical_features, drop_first=True)

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42, stratify=y)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
                        
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix

def evaluate_model(model, X_test, y_test, threshold=0.5, name="Model"):
    probs = model.predict_proba(X_test)[:, 1]
    preds = (probs >= threshold).astype(int)
    
    print(f"\n{'='*20} {name} {'='*20}")
    print(classification_report(y_test, preds))
    
    cm = confusion_matrix(y_test, preds)
    plt.figure(figsize=(4,3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='RdPu')
    plt.title(f"Confusion Matrix: {name}")
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

lr = LogisticRegression(max_iter=1000)
lr.fit(X_train_scaled, y_train)
evaluate_model(lr, X_test_scaled, y_test, threshold=0.3, name="Logistic Regression")

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_scaled, y_train) 
evaluate_model(rf, X_test_scaled, y_test, threshold=0.3, name="Random Forest")
```

```python
importances = pd.Series(rf.feature_importances_, index=X_train.columns)
importances.nlargest(10).plot(kind='barh', color='skyblue', figsize=(8,6))
plt.title("Top 10 Features Driving Credit Default")
plt.gca().invert_yaxis() 
plt.show()
```
